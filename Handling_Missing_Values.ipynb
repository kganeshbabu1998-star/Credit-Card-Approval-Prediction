```python
import pandas as pd
import numpy as np

# Load Dataset
df = pd.read_csv("application_record.csv.gz", compression="gzip")

print("="*60)
print("APPLICATION RECORD DATASET")
print("="*60)

# Display First 5 Rows
print("\nFirst 5 Rows:")
print(df.head())

# Dataset Information
print("\nDataset Information:")
df.info()

# Dataset Shape
print("\nDataset Shape:", df.shape)

# Missing Values Count
print("\nMissing Values Count:")
print(df.isnull().sum())

# Missing Values Percentage
print("\nMissing Values Percentage:")
missing_percentage = (df.isnull().sum()/len(df))*100
print(missing_percentage)

# Display Columns Having Missing Values
print("\nColumns with Missing Values:")
print(missing_percentage[missing_percentage > 0])

# Remove OCCUPATION_TYPE Column
df.drop(columns=['OCCUPATION_TYPE'], inplace=True)

print("\n'OCCUPATION_TYPE' column removed successfully.")

# Verify Missing Values
print("\nMissing Values After Cleaning:")
print(df.isnull().sum())

# Save Cleaned Dataset
df.to_csv("application_record_cleaned.csv", index=False)

print("\nCleaned dataset saved as application_record_cleaned.csv")
print("\nStory 2: Handling Missing Values Completed Successfully!")
```
