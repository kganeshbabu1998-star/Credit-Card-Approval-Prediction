```python
import pandas as pd
```

```python
application_df = pd.read_csv("application_record.csv")
credit_df = pd.read_csv("credit_record.csv")
```

```python
application_df.head()
```

```python
credit_df.head()
```

```python
print(application_df.duplicated().sum())
print(credit_df.duplicated().sum())
```

```python
application_df.drop_duplicates(inplace=True)
credit_df.drop_duplicates(inplace=True)
```

```python
print(application_df.duplicated().sum())
print(credit_df.duplicated().sum())
```

```python
application_df.to_csv("application_record_clean.csv", index=False)
credit_df.to_csv("credit_record_clean.csv", index=False)
```
